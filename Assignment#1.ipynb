```python
import numpy as num
m=data['diagnosis'].value_counts().idxmax()
data['diagnosis'].replace(num.nan,m,inplace=True)
data['diagnosis'].isnull().sum()
```

```python
m=data['subtlety'].value_counts().idxmax()
data['subtlety'].replace(num.nan,m,inplace=True)
m=data['size'].value_counts().idxmax()
data['size'].replace(num.nan,m,inplace=True)
m=data['age'].value_counts().idxmax()
data['age'].replace(num.nan,m,inplace=True)
m=data['x'].value_counts().idxmax()
data['x'].replace(num.nan,m,inplace=True)
m=data['y'].value_counts().idxmax()
data['y'].replace(num.nan,m,inplace=True)
m=data['position'].value_counts().idxmax()
data['position'].replace(num.nan,m,inplace=True)
data.replace("non-nodule",num.nan,inplace = True)
m=data['state'].value_counts().idxmax()
data['state'].replace(num.nan,m,inplace=True)
data.replace("unknown",num.nan,inplace = True)
m = data['diagnosis'].value_counts().idxmax()
data['diagnosis'].replace(num.nan,m, inplace = True)
```
